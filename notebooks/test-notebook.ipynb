```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

# we will generate noisy line data to fit a linear model to

```python
x = np.linspace(0, 1, 101)

nu_radius = 20
nu = (np.random.rand(101) * 2 - 1) * nu_radius

m = 500
b = 500

y = m * x + b
yn = y + nu

plt.plot(x, y, label="y")
plt.plot(x, yn, label="noisy y")

plt.legend()
```

# fit a linear model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

X = x.reshape((-1, 1))  # sklearn expects multidimensional features by default

yp = model.fit(X, yn).predict(X)


plt.plot(x, y, label="y")
plt.plot(x, yn, label="noisy y")
plt.scatter(x, yp, label="linear fit noisy y", c="g", s=5)

plt.legend()
```

```python
(m_est,) = model.coef_
b_est = model.intercept_

print(f"m: {m}, m_estimate: {m_est}")
print(f"b: {b}, b_estimate: {b_est}")
```
